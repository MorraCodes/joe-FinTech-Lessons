## Statistical describtion.
- count: The number of elements in the column.
- mean: The average value of the column.
- std: The standard deviation of the values in the column—that is, how spread out the values are (which indicates the  distribution).
- min: The minimum value in the column.
- 25%, 50%, 75%: The values that occupy each respective percentile (which also indicate the distribution). Note that the value that occupies the 50% percentile isn’t the same as the mean, although it’s close.
- max: The maximum value in the column.

## Standard Deviation.
 the standard deviation is a measure of the amount of variation or dispersion of a set of values. [1] A low standard deviation indicates that the values tend to be close to the mean (also called the expected value) of the set, while a high standard deviation indicates that the values are spread out over a wider range. Standard deviation may be abbreviated SD, and is most commonly represented in mathematical texts and equations by the lower case Greek letter sigma `σ`, for the population standard deviation, or the Latin letter s, for the sample standard deviation. [2]

## Variance.
variance is the expectation of the squared deviation of a random variable from its mean. Variance is a measure of dispersion, meaning it is a measure of how far a set of numbers is spread out from their average value.


# SD vs variance.
The standard deviation is expressed in the same units as the mean is, whereas the variance is expressed in squared units, but for looking at a distribution, you can use either just so long as you are clear about what you are using. For example, a Normal distribution with mean = 10 and sd = 3 is exactly the same thing as a Normal distribution with mean = 10 and variance = 9.

To get all the statistics `data` for our data we use `.describe()`

Let’s look at an example. Consider a DataFrame named returns_df that has three columns for Stock A, Stock B, and Stock C. The DataFrame has five rows for each day of the first full week of trading in 2020. To call the describe function on this DataFrame, we use the following code:
`returns_df.describe`



The Pandas describe function always outputs the same list of summary statistics:

    count: The number of elements in the column.
    mean: The average value of the column.
    std: The standard deviation of the values in the column—that is, how spread out the values are (which indicates the distribution).
    min: The minimum value in the column.
    25%, 50%, 75%: The values that occupy each respective percentile (which also indicate the distribution). Note that the value that occupies the 50% percentile isn’t the same as the mean, although it’s close.
    max: The maximum value in the column.

        - A histogram is an approximate representation of the distribution of numerical data.
        -  box plot: This type of visualization displays information about the distribution, or spread, of data. Box plots highlight the mean values and the minimum and maximum values. We call the latter two values outliers, because those data points lie far away from the mean and the main body of distribution. `We use a histogram (Links to an external site.) to illustrate how often values appear in the DataFrame (that is, the frequency of those values). Pandas dynamically groups the values into bins, which display along the x-axis. The number of elements in each bin displays along the y-axis. The result is similar to a bar chart. Analysts often use a histogram together with summary statistics to get a quick view of the spread of the elements in a DataFrame.`
        
        A Pandas box plot  (Links to an external site.) is a visual representation of the spread of the plotted data.

    The box displays the mean value and includes the range where we find the middle quartiles of data (that is, the largest concentration of data points).

    The whiskers of the box plot represent the outer quartiles of data, with the length of the whiskers illustrating how spread out the data points are.

    The edges of the plot illustrate the minimum and maximum values of the data.
